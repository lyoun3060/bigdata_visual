```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#차트에 한글 안깨지게
from matplotlib import rc, font_manager #(rc = objective, font_manager = method)
fontname = "C:/Windows/Fonts/gulim.ttc"
font_name = font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)

#마이너스 기호 안깨지게
plt.rcParams['axes.unicode_minus']=False
# Matplotlib 라이브러리에서 그림(figure)의 크기를 설정하는 코드
plt.rcParams['figure.figsize']=(6,3)
```

```python
#seaborn 내장에 있는 라이브러리중 titanic의 내용
titanic = sns.load_dataset('titanic')
titanic
```

```python
#seaborn 내장에 있는 라이브러리들을 보여줌
sns.get_dataset_names()
```

# 2. relplot
    - 두 columns간의 상관관계를 보여줌

```python
#sns.relplot(x='x축을 사용할 필드', y='y축을 사용할 피드', data='dataFrame')
print(tips['total_bill'].corr(tips['tip']))
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette="bright", style='day', size='size')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, palette="Pastel2_r", hue='smoker', style='day', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', col='day', row='sex', data=tips)
```

```python
sns.get_dataset_names()
taxis=sns.load_dataset('taxis')
taxis
```

```python
sns.relplot(x='total', y='tip', row='payment', col='color', hue='pickup_borough', data=taxis)
```

```python
#1~500까지 0~1사이의 값을 누적한 값들을 차트로 보여주는것
df= pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
sns.relplot(x='time', y='value', data=df, kind='line')
```

```python
fmri=sns.load_dataset('fmri')
fmri
#같은값이 여러개(여기서는 timepoint의 값)인 경우 
#최소, 최대 값을 연한선으로 보여주고 해당 값의 평균을 보여줌
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=100)
```

```python
# sns.distplot() : 히스토그램, 라인그래프, kde그래프
# 디폴트 => (러그(x), 히스트그래프(o), kde그래프(o))
x=np.random.randn(100)
x
# sns.distplot(x, rug=False, hist=True, kde=True)
sns.histplot(data=x, kde=True)
sns.rugplot(x=x)
plt.show()
```

# heatmap
 - 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지 위에 열분포형태의 비주얼한 그래픽 출력 가능

```python
data=np.random.rand(5,6)
data

sns.heatmap(data, annot=True)
```

```python
print(tips)
pivot=tips.pivot_table(index='day', columns='size', values='tip')
print(pivot)

sns.heatmap(pivot, cmap='Reds', annot=True)
```

```python
titanic_corr= titanic.corr()

sns.heatmap(titanic_corr, annot=True, cmap='plasma')
```
