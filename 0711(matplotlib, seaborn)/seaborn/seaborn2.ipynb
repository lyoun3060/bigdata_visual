```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
#차트에 한글 안깨지게
from matplotlib import rc, font_manager #(rc = objective, font_manager = method)
fontname = "C:/Windows/Fonts/gulim.ttc"
font_name = font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)

#마이너스 기호 안깨지게
plt.rcParams['axes.unicode_minus']=False
# Matplotlib 라이브러리에서 그림(figure)의 크기를 설정하는 코드
plt.rcParams['figure.figsize']=(6,3)
```

```python
tips=sns.load_dataset('tips')
tips.head() #처음 데이터 5개만 보기
tips.tail() #끝의 데이터 5개만 보기
```

# catplot
 - 범주를 나누어서 그려줌
 - 참고사이트 : https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
#kind='violin'사용할때는 split을 많이사용
sns.catplot(x='sex', y='total_bill', data=tips, kind='violin', hue='smoker', split=True, inner='stick')
```

```python
# Box plot
#주요값들이 어디에 몰려있는지 볼때 많이 사용
sns.catplot(x='day', y='total_bill', data=tips, kind='bar', hue='sex')
```
