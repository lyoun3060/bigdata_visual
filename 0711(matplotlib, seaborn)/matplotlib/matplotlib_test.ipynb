```python
#rand(변수) = 0~1사이에 (변수)번  랜덤하게 실수발생시키는것
data1=np.random.rand(50)
#randn(변수) =  -1~0~1사이에 (변수)번 랜덤하게 실수발생시키는것
data2 = np.random.randn(50)
#randint(변수, 변수1, 변수2) = 변수~변수1사이에 변수2번만큼 정수를 랜덤하게 생성시키는것
data3 = np.random.randint(1, 10, 5)
#rand(변수, 변수2) = 행(변수)개 열(변수2)개를 만드는것
data4 = np.random.rand(10,4)



df=pd.DataFrame(data4, columns=['A', 'B', 'C', 'D'])
df.plot()
```

```python
#차트같은거 쓸때 한글쓰면 깨짐현상 있음
#한글설정을 따로 해줘야함
from matplotlib import rc, font_manager #(rc = objective, font_manager = method)
fontname = "C:/Windows/Fonts/gulim.ttc"
font_name = font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)

#시리즈 형태의 데이터 입력
sri=pd.Series(np.random.rand(5), index=['A','B','C','D','E'], name='파이차트')

sri.plot(kind='pie')

print(sri)
#차트 안에 값을 표기하고자 할때, 실제 데이터와 다른 값(%로 변경)
sri.plot(kind='pie',autopct='%.2f')
```

```python
#마이너스 부호 깨지지 않게 설정
plt.rcParams['axes.unicode_minus']=False

#normal(중앙값, 표준편차(옆으로 벌어지는 갯수), 갯수)
s1=pd.Series(np.random.normal(0,100, size=200))
s1.hist(bins=50)
```

```python
x1= np.random.normal(1, 1, size=(100,1))
x2= np.random.normal(-2, 4, size=(100,1))
x = np.concatenate((x1, x2), axis=1)
df= pd.DataFrame(x, columns=['X1', 'X2'])
df['X1'].corr(df['X2'])
```

```python
plt.scatter(x1, x2)
```
