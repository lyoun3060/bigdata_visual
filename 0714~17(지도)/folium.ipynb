```python
from folium import plugins
import numpy as np
n=100
data=np.array([np.random.uniform(low=35, high=38, size=n), np.random.uniform(low=126, high=128, size=n)]).T
data
```

```python
popups=[str(i) for i in range(n)]
m=folium.Map(location=[latitude, longtitude], zoom_start=13)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m.save('data/cluster.html')
m
```

```python
#JSON으로 folium에 표기하기

import requests
import json
r= requests.get("https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json")
c=r.content
print(c)
seoul_geo = json.loads(c)
seoul_geo

```

```python
m=folium.Map(location=[latitude, longtitude], zoom_start=12)
folium.GeoJson(seoul_geo, name='지역구').add_to(m)
m
```

```python
cb = pd.read_csv('data/CoffeeBean.csv')
cb
```

```python
add=pd.DataFrame(cb.address.apply(lambda v : v.split()[:2]).to_list(),columns=['시도', '구군'])
add
#서울시, 서울, 서울특별시 등 으로 나옴 => 서울시로 통일 시켜줘야함

#바꾸고 싶은거 있을때 aliases를 씀
# addr_aliases={'서울' : '서울특별시', '서울시' : '서울특별시'}
# add['시도']=add['시도'].apply(lambda v: addr_aliases.get(v, v))
# add['시도'].unique()
```

```python
addr2 = add.loc[add['시도']=='서울특별시',:]
addr2
```

```python
addr_group=addr2.groupby('구군')['시도'].count()
addr_group
```

```python
m.choropleth(geo_data=seoul_geo,
             data=addr_group,
             fill_color = 'Blues',
             fill_opacity=0.7,
             line_opacity=0.2,
             key_on='properties.name',
             legend_name='지역구별 커피 업종수')
m
```
